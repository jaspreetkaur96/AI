```python
import pandas as pd
df = pd.read_csv("ai.csv")
df
```

```python
df.drop('Day',axis=1,inplace=True)
df
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
df['Outlook']=le.fit_transform(df['Outlook'])
df['Temp']=le.fit_transform(df['Temp'])
df['Humidity']=le.fit_transform(df['Humidity'])
df['Wind']=le.fit_transform(df['Wind'])
df['Tennis']=le.fit_transform(df['Tennis'])
```

```python
df
```

```python
X = df[["Outlook","Temp","Humidity","Wind"]]
y = df[["Tennis"]]
print(X)
print(y)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2) # To cluster on the basis of number of users.
kmeans.fit(X)
```

```python
prediction = kmeans.predict(X)
prediction
```

```python
from sklearn.metrics import accuracy_score
kmeans = KMeans(n_clusters=2) # Wind,Overcast,Humidity
kmeans.fit(X)
prediction_2 = kmeans.predict(X)
print("K=2",accuracy_score(prediction_2, y,))
```
